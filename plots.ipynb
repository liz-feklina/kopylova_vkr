## Распределение частей речи

```python
POS_ms_ict = df_ict.groupby(by=['POS_ms'])['POS_ms'].count()
POS_ms_not_ict = df_not_ict.groupby(by=['POS_ms'])['POS_ms'].count()

POS_ms_ict.sort_index()
```

```python
pos_distr = pandas.concat([POS_ms_ict.rename('ict'), POS_ms_not_ict.rename('not ict')], axis=1)
```

```python
def percentage(a, b):
    return round(a/(a+b), 2)

pos_distr['percentage of ict'] = pos_distr.apply(lambda x: percentage(x['ict'], x['not ict']), axis=1)
```

```python
pos_distr.sort_values(by='percentage of ict')
```

## Частотные слова

```python
freq_words_ict = df_ict.groupby(by=['word'])['word'].count()
freq_words_not_ict = df_not_ict.groupby(by=['word'])['word'].count()

freq_words = pandas.concat([freq_words_ict.rename('ict'), freq_words_not_ict.rename('not ict')], axis=1)
```

```python
freq_words = freq_words.fillna(0)
freq_words['ict'] = freq_words['ict'].astype(int)
freq_words['not ict'] = freq_words['not ict'].astype(int)
```

```python
freq_words['total number'] = freq_words.apply(lambda x: (x['ict'] + x['not ict']), axis=1)
freq_words['percentage of ict'] = freq_words.apply(lambda x: percentage(x['ict'], x['not ict']), axis=1)
```

```python
freq_words = freq_words[freq_words['total number'] > 1000]
```
