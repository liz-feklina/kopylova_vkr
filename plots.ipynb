```python
#df = pandas.read_csv('mono_df_all.csv', index_col=0, quotechar='&')
df = pandas.read_csv('/content/drive/MyDrive/diplom/mono_df_all.csv', index_col=0, quotechar='&')
```

```python
ict_count = df.groupby(by=['ict'])['ict'].count()
```

```python
df_ict = df[df['ict'] == True] 
df_not_ict = df[df['ict'] == False]
```

## Фонетическая структура

```python
sylls_ict = df_ict.groupby(by=['syll_type'])['syll_type'].count()
```

```python
print('Распределение типов фонетической структуры односложных слов, отмеченных иктом')
print(f'Количество закрытых слогов: {sylls_ict.values[0]} ({round(100*sylls_ict.values[0]/(sylls_ict.values[0] + sylls_ict.values[1]), 2)}%)')
print(f'Количество открытых слогов: {sylls_ict.values[1]} ({round(100*sylls_ict.values[1]/(sylls_ict.values[0] + sylls_ict.values[1]), 2)}%)')
```

```python
sylls_not_ict = df_not_ict.groupby(by=['syll_type'])['syll_type'].count()
```

```python
print('Распределение типов фонетической структуры односложных слов, не отмеченных иктом')
print(f'Количество закрытых слогов: {sylls_not_ict.values[0]} ({round(100*sylls_not_ict.values[0]/(sylls_not_ict.values[0] + sylls_not_ict.values[1]), 2)}%)')
print(f'Количество открытых слогов: {sylls_not_ict.values[1]} ({round(100*sylls_not_ict.values[1]/(sylls_not_ict.values[0] + sylls_not_ict.values[1]), 2)}%)')
```

```python
print('Распределение закрытых слогов:',
      f'{sylls_not_ict.values[0]} ({round(100*sylls_not_ict.values[0]/(sylls_not_ict.values[0] + sylls_ict.values[0]), 2)}%) среди слов, не отмеченных иктом;',
      f'{sylls_ict.values[0]} ({round(100*sylls_ict.values[0]/(sylls_not_ict.values[0] + sylls_ict.values[0]), 2)}%) среди слов, отмеченных иктом\n',
      sep = '\n')
print('Распределение открытых слогов:',
      f'{sylls_not_ict.values[1]} ({round(100*sylls_not_ict.values[1]/(sylls_not_ict.values[1] + sylls_ict.values[1]), 2)}%) среди слов, не отмеченных иктом;',
      f'{sylls_ict.values[1]} ({round(100*sylls_ict.values[1]/(sylls_not_ict.values[1] + sylls_ict.values[1]), 2)}%) среди слов, отмеченных иктом\n',
      sep = '\n')
```

```python
# построить график
x_tags = ['closed', 'closed', 'open', 'open']
y_tags =[38, 62, 70, 30]
i_tags = ['not ict', 'ict', 'not ict', 'ict']

syll_data = pandas.DataFrame(list(zip(x_tags, y_tags, i_tags)),
               columns =['syllable type', 'percentage', 'ict'])

fig = px.bar(syll_data, x="syllable type", y="percentage", color="ict", title="syllable structure")

fig.update_layout(
    autosize=False,
    width=500,
    height=500,
)

fig.show()
```
